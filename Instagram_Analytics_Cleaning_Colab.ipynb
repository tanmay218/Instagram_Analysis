```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Instagram_Analytics_Unclean.csv')

print("Original shape:", df.shape)
df.head()
```

```python
print(df.info())

print("\nMissing values:\n", df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
text_cols = ['post_id', 'media_type', 'traffic_source', 'content_category', 'upload_date']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

df.head()
```

```python
df.replace(['unknown', 'N/A', '-', 'missing', '%error', 'not available', 'nan', 'None'], np.nan, inplace=True)

df.isnull().sum()
```

```python
# media_type
df['media_type'] = df['media_type'].replace({
    'photo': 'Photo',
    'VIDEO': 'Video',
    'video': 'Video',
    ' reel ': 'Reel',
    'reel': 'Reel',
    'Carousal': 'Carousel'
})

# traffic_source
df['traffic_source'] = df['traffic_source'].replace({
    'home feed': 'Home Feed',
    'EXPLORE': 'Explore',
    ' profile ': 'Profile',
    'HashTags': 'Hashtags',
    'reels feed': 'Reels Feed',
    'external ': 'External'
})

# content_category
df['content_category'] = df['content_category'].replace({
    'tech': 'Technology',
    ' fitness ': 'Fitness',
    'beauty ': 'Beauty',
    'MUSIC': 'Music',
    ' travel': 'Travel'
})

df.head()
```

```python
numeric_cols = [
    'likes', 'comments', 'shares', 'saves', 'reach', 'impressions',
    'caption_length', 'hashtags_count', 'followers_gained', 'engagement_rate'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dtypes
```

```python
df['upload_date'] = pd.to_datetime(df['upload_date'], errors='coerce', dayfirst=True)

df.head()
```

```python
print("Duplicates before:", df.duplicated().sum())

df = df.drop_duplicates()

print("After removal:", df.duplicated().sum())
print("New shape:", df.shape)
```

```python
for col in numeric_cols:
    df.loc[df[col] < 0, col] = np.nan

df.describe()
```

```python
def treat_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    data.loc[(data[column] < lower) | (data[column] > upper), column] = np.nan
    return data

for col in numeric_cols:
    df = treat_outliers_iqr(df, col)

df.describe()
```

```python
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

df.isnull().sum()
```

```python
cat_cols = ['media_type', 'traffic_source', 'content_category']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

```python
df['post_id'] = df['post_id'].replace('nan', np.nan)
df['post_id'] = df['post_id'].fillna('Unknown_ID')

df.head()
```

```python
valid_dates = df['upload_date'].dropna().sort_values()

if len(valid_dates) > 0:
    middle_date = valid_dates.iloc[len(valid_dates)//2]
    df['upload_date'] = df['upload_date'].fillna(middle_date)

df.head()
```

```python
df['engagement_rate'] = df['engagement_rate'].round(2)

df['upload_date'] = df['upload_date'].dt.strftime('%Y-%m-%d')

df.head()
```

```python
print("Final shape:", df.shape)

print("\nMissing values:\n", df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

print("\nData types:\n", df.dtypes)
```

```python
df.to_csv('/content/Instagram_Analytics_Cleaned.csv', index=False)

print("Cleaned dataset saved successfully.")
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Fix traffic_source
df['traffic_source'] = df['traffic_source'].astype(str).str.strip().replace({
    'profile': 'Profile',
    'external': 'External'
})

# Fix content_category
df['content_category'] = df['content_category'].astype(str).str.strip().replace({
    'beauty': 'Beauty',
    'fitness': 'Fitness',
    'travel': 'Travel'
})

# Convert shares to integer
df['shares'] = df['shares'].round().astype(int)

# Convert other count columns to integer
count_cols = [
    'likes', 'comments', 'saves', 'reach', 'impressions',
    'caption_length', 'hashtags_count', 'followers_gained'
]

for col in count_cols:
    df[col] = df[col].round().astype(int)

# Round engagement rate
df['engagement_rate'] = df['engagement_rate'].round(2)

# Final validation
print("Duplicate rows:", df.duplicated().sum())

print("\nTraffic Source values:")
print(sorted(df['traffic_source'].unique()))

print("\nContent Category values:")
print(sorted(df['content_category'].unique()))

print("\nData types:")
print(df.dtypes)

# Save final cleaned dataset
df.to_csv('/content/Instagram_Analytics_Final_Clean.csv', index=False)

print("\nFinal clean dataset saved successfully.")
```
